# Câu 15
Viết thuật toán phương pháp lặp đơn cho hệ phương trình phi tuyến

## Thuật toán lặp đơn giải hệ phương trình phi tuyến

### Bài toán
Giải hệ $\mathbf{x} = \boldsymbol{\varphi}(\mathbf{x})$, tức là:
$$x_i = \varphi_i(x_1, x_2, \ldots, x_n), \quad i = 1, 2, \ldots, n$$

### Điều kiện hội tụ
$$\|J_{\varphi}(x)\|_{\infty} = \max_i \sum_j \left|\frac{\partial \varphi_i}{\partial x_j}\right| \leq q < 1, \quad \forall x \in D$$

### Thuật toán (sai số hậu nghiệm)

**Input:**
- $x^{(0)} = (x_1^{(0)}, x_2^{(0)}, \ldots, x_n^{(0)})$ — điểm xuất phát
- $\varepsilon$ — sai số yêu cầu
- $q$ — hằng số hội tụ

**Output:** $x^* \approx \xi$ — nghiệm gần đúng

```
BEGIN
  k ← 0
  REPEAT
    FOR i = 1 TO n DO
      x_new[i] ← φᵢ(x0[1], x0[2], ..., x0[n])
    END FOR
    err ← ||x_new - x0||∞
    x0  ← x_new
    k   ← k + 1
  UNTIL err < ε*(1-q)/q
  PRINT x0, k
END
```

### Công thức dừng

Dừng khi: $\|x^{(k)} - x^{(k-1)}\|_{\infty} < \varepsilon\cdot\dfrac{1-q}{q}$

Khi đó sai số thực thỏa: $\|x^{(k)} - \xi\| \leq \dfrac{q}{1-q} \cdot \mathrm{err} < \varepsilon$

### Ước lượng số bước lặp cần thiết (tiên nghiệm)

$$k \geq \frac{\ln\!\left(\dfrac{\varepsilon(1-q)}{\|x^{(1)} - x^{(0)}\|_{\infty}}\right)}{\ln q}$$

````python
import numpy as np

def lap_don(phi, x0, q, eps=10**(-6), max_iter=1000):
    x = np.array(x0, dtype=float)
    n = len(x)
    header = "  k  |" + "".join(f"  x{i+1}{'':>10}" for i in range(n)) + "|    sai số"
    print(header)
    print("-" * len(header))

    for k in range(1, max_iter + 1):
        x_new = np.array(phi(x))
        err = np.max(np.abs(x_new - x))
        x = x_new
        row = f" {k:3d} |" + "".join(f"  {x[i]:14.8f}" for i in range(n)) + f"| {err:.10f}"
        print(row)
        if err < eps * (1 - q) / q:
            return x, k
    raise RuntimeError(f"Không hội tụ sau {max_iter} vòng lặp")


# --- Ví dụ ---
# Hệ:  x1 = (cos(x2) + 1) / 3
#       x2 = (sin(x1) + 2) / 4
def phi(x):
    return [(np.cos(x[1]) + 1) / 3,
            (np.sin(x[0]) + 2) / 4]

x0 = [0.5, 0.5]
q  = 0.5

x_star, k = lap_don(phi, x0, q, eps=10**(-6))
print(f"\nNghiệm: x1 = {x_star[0]:.8f},  x2 = {x_star[1]:.8f}")
print(f"Số vòng lặp: {k}")
````

````python
# --- Ví dụ 3 chiều ---

def phi3(x):
    return [(13 - x[1]**2 + 4*x[2]) / 15,
            (11+x[2] - x[0]**2) / 10,
            (22 + x[1]**3) / 25 ]

x0 = [0.0, 0.0, 0.0]
q  = 17/30

x_star, k = lap_don(phi3, x0, q, eps=10**(-6))
print(f"\nNghiệm: x1={x_star[0]:.8f}, x2={x_star[1]:.8f}, x3={x_star[2]:.8f}")
print(f"Số vòng lặp: {k}")
````

````python
# --- Ví dụ 3 chiều ---

def phi3(x):
    return [(13 - x[1] ** 2 + 4*x[2]) / 15,
            (np.sqrt(x[0]**2 + 0.3125)) / 25 - 0.03,
            (-np.exp(-x[0] * x[1])) / 20 - (10*np.pi - 3)/60]

x0 = [0.0, 0.0, 0.0]
q  = 2/3

x_star, k = lap_don(phi3, x0, q, eps=10**(-6))
print(f"\nNghiệm: x1={x_star[0]:.8f}, x2={x_star[1]:.8f}, x3={x_star[2]:.8f}")
print(f"Số vòng lặp: {k}")
````

# Câu 16
Viết thuật toán tìm nghiệm xấp xỉ hệ phương trình phi tuyến bằng phương pháp lặp Newton

## Phương pháp Newton giải hệ phương trình phi tuyến

### Bài toán
Tìm nghiệm $\mathbf{x}^*$ của hệ $n$ phương trình phi tuyến:

$$\mathbf{F}(\mathbf{x}) = \mathbf{0}$$

tức là:

$$\begin{cases} f_1(x_1, x_2, \ldots, x_n) = 0 \\ f_2(x_1, x_2, \ldots, x_n) = 0 \\ \vdots \\ f_n(x_1, x_2, \ldots, x_n) = 0 \end{cases}$$

### Điều kiện áp dụng
1. $\mathbf{F}$ khả vi liên tục bậc 2 trên miền $D$
2. $J_F(\boldsymbol{\xi})$ khả nghịch, tức $\det(J_F) \neq 0$ tại nghiệm $\boldsymbol{\xi}$
3. $\mathbf{x}^{(0)}$ đủ gần nghiệm $\boldsymbol{\xi}$

### Ma trận Jacobi

$$J_F(\mathbf{x}) = \begin{pmatrix} \dfrac{\partial f_1}{\partial x_1} & \cdots & \dfrac{\partial f_1}{\partial x_n} \\[10pt] \vdots & \ddots & \vdots \\[10pt] \dfrac{\partial f_n}{\partial x_1} & \cdots & \dfrac{\partial f_n}{\partial x_n} \end{pmatrix}$$

### Công thức lặp

$$J_F(\mathbf{x}^{(k)}) \cdot \boldsymbol{\delta}^{(k)} = -\mathbf{F}(\mathbf{x}^{(k)}), \qquad \mathbf{x}^{(k+1)} = \mathbf{x}^{(k)} + \boldsymbol{\delta}^{(k)}$$

> Không tính $J_F^{-1}$ trực tiếp — dùng khử Gauss hoặc LU để giải hệ.

---

### Thuật toán 1 — Dừng sau $N$ bước

**Input:**
- $x^{(0)} = (x_1^{(0)}, \ldots, x_n^{(0)})$ — điểm xuất phát
- $N$ — số bước lặp

**Output:** $x^{(N)}$ — nghiệm sau $N$ bước

```
BEGIN
  FOR k = 0 TO N-1 DO
    Tính F(x^(k))  và  J_F(x^(k))
    Giải hệ:  J_F(x^(k)) · δ = −F(x^(k))
    x^(k+1) ← x^(k) + δ
    In ra:  k+1,  x^(k+1),  ||δ||∞
  END FOR
  PRINT x^(N)
END
```

---

### Thuật toán 2 — Dừng theo $\varepsilon$

**Input:**
- $x^{(0)} = (x_1^{(0)}, \ldots, x_n^{(0)})$ — điểm xuất phát
- $\varepsilon$ — sai số yêu cầu

**Output:** $x^*$ — nghiệm gần đúng

```
BEGIN
  k ← 0
  REPEAT
    Tính F(x^(k))  và  J_F(x^(k))
    Giải hệ:  J_F(x^(k)) · δ = −F(x^(k))
    x^(k+1) ← x^(k) + δ
    k       ← k + 1
    In ra:  k,  x^(k),  ||δ||∞
  UNTIL ||δ||∞ < ε
  PRINT x^(k)
END
```

---

### Ước lượng sai số

Hội tụ bậc 2 — số chữ số đúng nhân đôi sau mỗi bước:

$$\|\mathbf{e}^{(k+1)}\| \leq C \cdot \|\mathbf{e}^{(k)}\|^2, \qquad C = \frac{1}{2}\|J_F(\boldsymbol{\xi})^{-1}\| \cdot \|F''(\boldsymbol{\xi})\|$$

| Bước $k$ | 0 | 1 | 2 | 3 | 4 |
|:---:|:---:|:---:|:---:|:---:|:---:|
| Sai số | $10^{-1}$ | $10^{-2}$ | $10^{-4}$ | $10^{-8}$ | $10^{-16}$ |

---

### Các lưu ý

| | Thuật toán 1 ($N$ bước) | Thuật toán 2 ($\varepsilon$) |
|---|---|---|
| Dùng khi | Bài tập, trình bày tay | Lập trình thực tế |
| Biết sai số? | Không | Có |

- **Chọn $x^{(0)}$:** phải đủ gần nghiệm; có thể dùng lặp đơn vài bước để khởi tạo
- **$J_F$ suy biến:** nếu $\det(J_F) \approx 0$ thì $\boldsymbol{\delta}$ rất lớn, Newton mất ổn định
- **Nghiệm bội:** $J_F(\boldsymbol{\xi})$ suy biến → hội tụ bậc 1 thay vì bậc 2
- **Newton đơn giản:** tính $J_F$ một lần tại $x^{(0)}$, dùng lại → rẻ hơn nhưng hội tụ bậc 1

````python
import numpy as np

def _print_header(n):
    header = "  k  |" + "".join(f"  x{i+1}{'':>10}" for i in range(n)) + "|    ||δ||∞"
    print(header)
    print("-" * len(header))

def _print_row(k, x, err, n):
    print(f" {k:3d} |" + "".join(f"  {x[i]:14.8f}" for i in range(n)) + f"| {err:.10f}")


# ── Thuật toán 1: dừng sau N bước ──────────────────────────────────────────
def newton_N(F, J, x0, N):
    x = np.array(x0, dtype=float)
    n = len(x)
    _print_header(n)
    for k in range(N):
        delta = np.linalg.solve(J(x), -F(x))
        x = x + delta
        _print_row(k + 1, x, np.max(np.abs(delta)), n)
    print(f"\nNghiệm: {x}")


# ── Thuật toán 2: dừng theo ε ───────────────────────────────────────────────
def newton_eps(F, J, x0, eps=1e-8):
    x = np.array(x0, dtype=float)
    n, k = len(x), 0
    _print_header(n)
    while True:
        delta = np.linalg.solve(J(x), -F(x))
        x = x + delta
        k += 1
        err = np.max(np.abs(delta))
        _print_row(k, x, err, n)
        if err < eps:
            break
    print(f"\nNghiệm: {x}  (sau {k} bước)")


# ── Ví dụ ───────────────────────────────────────────────────────────────────
# Hệ:  x1² + x2² - 5 = 0
#       x1 - x2 - 1  = 0
def F(x):
    return np.array([x[0]**2 + x[1]**2 - 5,
                     x[0]  - x[1]  - 1])

def J(x):
    return np.array([[2*x[0], 2*x[1]],
                     [1,     -1     ]])

x0 = [1.0, 0.0]

print("=== Thuật toán 1 (N=4) ===")
newton_N(F, J, x0, N=4)

print("\n=== Thuật toán 2 (ε=1e-8) ===")
newton_eps(F, J, x0, eps=1e-8)
````

````python
def F(x):
    return np.array([4*x[0]**2 - 20*x[0] + 1/4 * x[1]**2 + 8,
                     1/2 * x[0] * x[1]**2 + 2*x[0] - 5*x[1] + 8])
def J(x):
    return np.array([[8*x[0] - 20, 1/2 * x[1]],
                     [1/2 * x[1]**2 + 2, x[0] * x[1] - 5]])

x0 = [0.0, 0.0]
newton_N(F,J,x0,N = 5 )
````

````python
def F3(x):
    return np.array([3*x[0] - np.cos(x[1]*x[2]) - 1/2,
                     4*x[0]**2 - 625*x[1]**2 + 2*x[1] - 1,
                     np.exp(-x[0]*x[1]) + 20*x[2] + (10*np.pi - 3)/3])
def J3(x):
    return np.array([[3, 0, 0],
                     [8*x[0], -1250*x[1] + 2, 0],
                     [-x[1]*np.exp(-x[0]*x[1]), -x[0]*np.exp(-x[0]*x[1]), 20]])
x0 = [0.0, 0.0, 0.0]
newton_N(F3,J3,x0,N = 5 )
````
