# Câu 7:
Viết thuật toán hoặc vẽ sơ đồ khối cho **phương pháp dây cung / tiếp tuyến** giải phương trình $f(x) = 0$ trên khoảng $(a, b)$ theo:
- Công thức sai số mục tiêu
- Công thức sai số theo hai lần xấp xỉ liên tiếp.

#### Thuật toán phương pháp dây cung theo công thức sai số mục tiêu

Input
$a, b, \epsilon, m$ ( $m$ là min của hàm số $|f'(x)|$ )

Các bước:

**Bước 1:** Chọn điểm xuất phát $x_0$

$$x_0 = \begin{cases} b & \text{nếu } a \text{ là điểm Fourier} \\ a & \text{nếu } b \text{ là điểm Fourier} \end{cases}$$

**Bước 2:** Tính điểm xấp xỉ tiếp theo

$$x_1 = x_0 - \frac{f(x_0)(x_0 - a)}{f(x_0) - f(a)}$$

**Bước 3:** Kiểm tra sai số

$$err = \frac{|f(x_1)|}{m}$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_0 = x_1, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_1 \end{cases}$$

Output:
$$x_1 \approx \text{nghiệm của } f(x)$$

---

#### Thuật toán phương pháp dây cung theo công thức sai số 2 lần xấp xỉ liên tiếp

Input
$a, b, \epsilon, M, m$

Các bước

**Bước 1:** Chọn điểm xuất phát $x_0$

$$x_0 = \begin{cases} b & \text{nếu } a \text{ là điểm Fourier} \\ a & \text{nếu } b \text{ là điểm Fourier} \end{cases}$$

**Bước 2:** Tính điểm xấp xỉ tiếp theo

$$x_1 = x_0 - \frac{f(x_0)(x_0 - a)}{f(x_0) - f(a)}$$

**Bước 3:** Kiểm tra sai số

$$err = \frac{M - m}{m} \cdot |x_1 - x_0|$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_0 = x_1, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_1 \end{cases}$$

Output
$$x_1 \approx \text{nghiệm của } f(x)$$

[!NOTE]
**Điều kiện hội tụ của phương pháp dây cung/tiếp tuyến** trên $[a, b]$:
1. $f(a) \cdot f(b) < 0$ — hàm trái dấu tại 2 đầu mút
2. $f'(x) \neq 0$ — hàm đơn điệu trên $[a, b]$
3. $f''(x)$ không đổi dấu trên $[a, b]$

---

#### Thuật toán phương pháp tiếp tuyến Newton

Input
$f(x),\ f'(x),\ x_0\ \text{(điểm Fourier)},\ \epsilon$

##### Cách 1: Sai số theo 2 lần xấp xỉ liên tiếp

**Bước 1:** Nhập $f(x),\ f'(x),\ x_0,\ \epsilon$

**Bước 2:** Tính

$$x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}, \quad err = |x_1 - x_0|$$

**Bước 3:** Kiểm tra sai số

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_0 = x_1, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_1 \end{cases}$$

---

##### Cách 2: Sai số theo công thức mục tiêu

> [!NOTE]
> Cần nhập thêm $M_2 = \max|f''(x)|$ và $m_1 = \min|f'(x)|$ trên $[a, b]$

**Bước 1:** Nhập $f(x),\ f'(x),\ x_0,\ \epsilon,\ M_2,\ m_1$

**Bước 2:** Tính

$$x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}, \quad err = \frac{M_2}{2m_1} \cdot |x_1 - x_0|^2$$

**Bước 3:** Kiểm tra sai số

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_0 = x_1, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_1 \end{cases}$$

---

Output
$$x_1 \approx \text{nghiệm của } f(x)$$

# Câu 8:
Viết thuật toán tính gần đúng hằng số $e$ (hằng số $\pi$) với sai số $\varepsilon$ cho trước.  
Áp dụng tính $e$ (tương ứng $\pi$) với **7 chữ số đáng tin** sau dấu phẩy bằng phương pháp dây cung / tiếp tuyến.


### Tính gần đúng số $e$ bằng phương pháp dây cung

Bài toán
Tìm nghiệm gần đúng của phương trình:
$$f(x) = \ln(x) - 1 = 0 \quad \text{trên } [2, 3]$$

Bước 1: Kiểm tra điều kiện hội tụ trên $[2, 3]$

$$f'(x) = \frac{1}{x} > 0 \quad \Rightarrow f'(x) \text{ không đổi dấu} \checkmark$$

$$f''(x) = -\frac{1}{x^2} < 0 \quad \Rightarrow f''(x) \text{ không đổi dấu} \checkmark$$

$$f(2) = \ln 2 - 1 < 0, \quad f(3) = \ln 3 - 1 > 0 \quad \Rightarrow f(2) \cdot f(3) < 0 \checkmark$$

$$f(2) \cdot f''(2) = (\ln 2 - 1) \cdot \left(-\frac{1}{4}\right) > 0 \quad \Rightarrow x = 2 \text{ là điểm Fourier} \checkmark$$

Bước 2: Áp dụng phương pháp dây cung

$$\epsilon = \frac{1}{2} \times 10^{-7}, \quad d = 2 \text{ (điểm Fourier, cố định)}, \quad x_0 = 3$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n) \cdot (x_n - d)}{f(x_n) - f(d)}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

Bước 3: Kiểm tra điều kiện dừng

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx e \end{cases}$$

Output
$$x_{n+1} \approx e$$

```python
#log_2(x) = np.log2(x); log_10(x) = np.log10(x) ;log_a(x) = np.log(x) / np.log(a)

def f(x):
    return np.log(x) - 1

def bisect_chord(f, d, x0, epsilon=0.5*10**(-7)):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0) * (x0 - d) / (f(x0) - f(d))
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

root = bisect_chord(f, d=2, x0=3)
print(f"\nNghiệm ≈ {root:.10f}")
```

### Tính gần đúng số $e$ bằng phương pháp tiếp tuyến Newton

Bài toán
Tìm nghiệm gần đúng của phương trình:
$$f(x) = \ln(x) - 1 = 0 \quad \text{trên } [2, 3]$$

Bước 1: Kiểm tra điều kiện hội tụ trên $[2, 3]$

$$f'(x) = \frac{1}{x} > 0 \quad \Rightarrow f'(x) \text{ không đổi dấu} \checkmark$$

$$f''(x) = -\frac{1}{x^2} < 0 \quad \Rightarrow f''(x) \text{ không đổi dấu} \checkmark$$

$$f(2) = \ln 2 - 1 < 0, \quad f(3) = \ln 3 - 1 > 0 \quad \Rightarrow f(2) \cdot f(3) < 0 \checkmark$$

$$f(2) \cdot f''(2) = (\ln 2 - 1) \cdot \left(-\frac{1}{4}\right) > 0 \quad \Rightarrow x_0 = 2 \text{ là điểm Fourier} \checkmark$$

Bước 2: Áp dụng phương pháp tiếp tuyến

$$\epsilon = \frac{1}{2} \times 10^{-7}, \quad x_0 = 2 \text{ (điểm Fourier, điểm xuất phát)}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

Bước 3: Kiểm tra điều kiện dừng

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại Bước 2} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx e \end{cases}$$

Output
$$x_{n+1} \approx e$$

```python
def f(x):
    return np.log(x) - 1
def df(x):
    return 1 / x

def newton(f, df, x0, epsilon=0.5*10**(-7)):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0) / df(x0)
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

root = newton(f, df, x0=3)
print(f"\nNghiệm ≈ {root:.10f}")
```

### Tính gần đúng số $\pi$ bằng phương pháp dây cung và tiếp tuyến Newton

Bài toán
Tìm nghiệm gần đúng của phương trình:
$$f(x) = \tan\left(\frac{x}{4}\right) - 1 = 0 \quad \text{trên } [2, 4]$$

$$\text{Nghiệm đúng: } \frac{x}{4} = \frac{\pi}{4} \Rightarrow x = \pi$$

Bước 1: Kiểm tra điều kiện hội tụ trên $[2, 4]$

$$f'(x) = \frac{1}{4}\sec^2\!\left(\frac{x}{4}\right) > 0 \quad \Rightarrow f'(x) \text{ không đổi dấu} \checkmark$$

$$f''(x) = \frac{1}{8}\sec^2\!\left(\frac{x}{4}\right)\tan\!\left(\frac{x}{4}\right) > 0 \quad \Rightarrow f''(x) \text{ không đổi dấu} \checkmark$$

$$f(2) = \tan(0.5) - 1 \approx -0.46 < 0, \quad f(4) = \tan(1) - 1 \approx 0.56 > 0 \quad \Rightarrow f(2) \cdot f(4) < 0 \checkmark$$

$$f(4) \cdot f''(4) > 0 \quad \Rightarrow x_0 = 4 \text{ là điểm Fourier} \checkmark$$

---

#### Phương pháp dây cung

$$d = 4 \text{ (điểm Fourier, cố định)}, \quad x_0 = 2, \quad \epsilon = \frac{1}{2} \times 10^{-7}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n) \cdot (x_n - d)}{f(x_n) - f(d)}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx \pi \end{cases}$$

---

#### Phương pháp tiếp tuyến Newton

$$x_0 = 4 \text{ (điểm Fourier)}, \quad \epsilon = \frac{1}{2} \times 10^{-7}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = x_n - \frac{\tan\!\left(\frac{x_n}{4}\right) - 1}{\frac{1}{4}\sec^2\!\left(\frac{x_n}{4}\right)}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx \pi \end{cases}$$

---

Output
$$x_{n+1} \approx \pi$$

```python
def f(x):
    return np.tan(x / 4) - 1

def df(x):
    return 1 / (4 * np.cos(x / 4)**2)

# Phương pháp dây cung
def day_cung(f, d, x0, epsilon=0.5*10**(-7)):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0) * (x0 - d) / (f(x0) - f(d))
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

# Phương pháp tiếp tuyến Newton
def newton(f, df, x0, epsilon=0.5*10**(-7)):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0) / df(x0)
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

print("=== Phương pháp dây cung ===")
r1 = day_cung(f, d=4, x0=2)
print(f"\nDây cung : π ≈ {r1:.7f}   |   Sai số so với π: {abs(r1 - np.pi):.10f}")

print("\n=== Phương pháp Newton ===")
r2 = newton(f, df, x0=4)
print(f"Newton   : π ≈ {r2:.7f}   |   Sai số so với π: {abs(r2 - np.pi):.15f}")
```

# Câu 9:
Viết thuật toán tính gần đúng $\sqrt[n]{a}$, $a \in \mathbb{R}$, $n \in \mathbb{N}$ với sai số $\varepsilon$ cho trước bằng phương pháp dây cung / tiếp tuyến.  
Áp dụng tính $\sqrt[3]{17}$ với **6 chữ số đáng tin**.


### Tính gần đúng $\sqrt[n]{a}$ bằng phương pháp dây cung và tiếp tuyến Newton

Bài toán
Tìm nghiệm gần đúng của phương trình:
$$f(x) = x^n - a = 0, \quad a \in \mathbb{R},\ n \in \mathbb{N}$$

Bước 1: Kiểm tra điều kiện hội tụ trên $[b_1, b_2]$

$$f'(x) = nx^{n-1} > 0 \quad \text{(với } x > 0\text{)} \quad \Rightarrow f'(x) \text{ không đổi dấu} \checkmark$$

$$f''(x) = n(n-1)x^{n-2} > 0 \quad \text{(với } x > 0, n \geq 2\text{)} \quad \Rightarrow f''(x) \text{ không đổi dấu} \checkmark$$

$$f(b_1) \cdot f(b_2) < 0 \quad \Rightarrow \text{chọn } b_1, b_2 \text{ sao cho } b_1^n < a < b_2^n \checkmark$$

$$f(b_2) \cdot f''(b_2) > 0 \quad \Rightarrow x_0 = b_2 \text{ là điểm Fourier} \checkmark$$

> [!NOTE]
> Thường chọn $b_1 = \lfloor\sqrt[n]{a}\rfloor$ và $b_2 = b_1 + 1$ để khoảng nhỏ nhất có thể

---

#### Phương pháp dây cung

$$d = b_2 \text{ (điểm Fourier, cố định)}, \quad x_0 = b_1, \quad \epsilon \text{ cho trước}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n) \cdot (x_n - d)}{f(x_n) - f(d)}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx \sqrt[n]{a} \end{cases}$$

---

#### Phương pháp tiếp tuyến Newton

$$x_0 = b_2 \text{ (điểm Fourier)}, \quad \epsilon \text{ cho trước}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = x_n - \frac{x_n^n - a}{n \cdot x_n^{n-1}} = \frac{(n-1)x_n^n + a}{n \cdot x_n^{n-1}}$$

**Sai số:**

$$err = |x_{n+1} - x_n|$$

$$\begin{cases} err \geq \epsilon & \rightarrow \text{đặt } x_n = x_{n+1}, \text{ quay lại} \\ err < \epsilon & \rightarrow \text{in ra } x_{n+1} \approx \sqrt[n]{a} \end{cases}$$

---

 Output
$$x_{n+1} \approx \sqrt[n]{a}$$

```python
import numpy as np

def tinh_can(a, n, epsilon=1/2*10**(-6)):
    def f(x):  return x**n - a
    def df(x): return n * x**(n-1)

    b1 = int(a**(1/n))
    b2 = b1 + 1

    header = f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}"
    sep = "-" * 72

    def day_cung():
        d, x0 = b2, b1
        print("  Phương pháp dây cung:")
        print(header)
        print(sep)
        k = 0
        while True:
            x1 = x0 - f(x0) * (x0 - d) / (f(x0) - f(d))
            k += 1
            print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
            if abs(x1 - x0) < epsilon:
                print(sep)
                return x1
            x0 = x1

    def newton():
        x0 = b2
        print("  Phương pháp Newton:")
        print(header)
        print(sep)
        k = 0
        while True:
            x1 = x0 - f(x0) / df(x0)
            k += 1
            print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
            if abs(x1 - x0) < epsilon:
                print(sep)
                return x1
            x0 = x1

    r1 = day_cung()
    r2 = newton()
    print(f"\nDây cung : {n}√{a} ≈ {r1:.10f}")
    print(f"Newton   : {n}√{a} ≈ {r2:.10f}")
    print(f"Chính xác:        {a**(1/n):.10f}")

# Ví dụ
tinh_can(a=17, n=3)   # ∛17
```

### Bài toán
$$f(x) = e^x - \cos 2x = 0 \quad \text{trên } [-0.6,\ -0.3]$$

Bước 1: Kiểm tra điều kiện hội tụ trên $[-0.6, -0.3]$

$$f(-0.6) = e^{-0.6} - \cos(-1.2) \approx 0.549 - 0.362 = 0.187 > 0$$

$$f(-0.3) = e^{-0.3} - \cos(-0.6) \approx 0.741 - 0.825 = -0.084 < 0$$

$$\Rightarrow f(-0.6) \cdot f(-0.3) < 0 \checkmark$$

$$f'(x) = e^x + 2\sin 2x \quad \text{không đổi dấu trên } [-0.6, -0.3] \checkmark$$

$$f''(x) = e^x + 4\cos 2x \quad \text{không đổi dấu trên } [-0.6, -0.3] \checkmark$$

$$f(-0.6) \cdot f''(-0.6) > 0 \quad \Rightarrow x_0 = -0.6 \text{ là điểm Fourier} \checkmark$$

$$\epsilon = \frac{1}{2} \times 10^{-8} \quad \text{(8 chữ số đáng tin)}$$

---

### Phương pháp dây cung

$$d = -0.6 \text{ (điểm Fourier, cố định)}, \quad x_0 = -0.3$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{f(x_n) \cdot (x_n - d)}{f(x_n) - f(d)}$$

$$\begin{cases} |x_{n+1} - x_n| \geq \epsilon & \rightarrow x_n = x_{n+1}, \text{ lặp tiếp} \\ |x_{n+1} - x_n| < \epsilon & \rightarrow \text{in ra } x_{n+1} \end{cases}$$

---

### Phương pháp tiếp tuyến Newton

$$x_0 = -0.6 \text{ (điểm Fourier)}$$

**Công thức lặp:**

$$x_{n+1} = x_n - \frac{e^{x_n} - \cos 2x_n}{e^{x_n} + 2\sin 2x_n}$$

$$\begin{cases} |x_{n+1} - x_n| \geq \epsilon & \rightarrow x_n = x_{n+1}, \text{ lặp tiếp} \\ |x_{n+1} - x_n| < \epsilon & \rightarrow \text{in ra } x_{n+1} \end{cases}$$

---

Output
$$x_{n+1} \approx \text{nghiệm âm lớn nhất của } e^x - \cos 2x = 0$$

```python
import numpy as np
def f(x):  return np.exp(x) - np.cos(2*x)
def df(x): return np.exp(x) + 2*np.sin(2*x)

a, b = -0.6, -0.3
epsilon = 0.5*10**(-8)

def day_cung(f, d, x0, epsilon):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0)*(x0 - d)/(f(x0) - f(d))
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

def newton(f, df, x0, epsilon):
    print(f"{'k':>4} | {'x_k':>14} | {'x_{k+1}':>14} | {'f(x_{k+1})':>14} | {'|x_{k+1}-x_k|':>14}")
    print("-" * 72)
    k = 0
    while True:
        x1 = x0 - f(x0)/df(x0)
        k += 1
        print(f"{k:>4} | {x0:>14.10f} | {x1:>14.10f} | {f(x1):>14.8f} | {abs(x1-x0):>14.8f}")
        if abs(x1 - x0) < epsilon:
            print("-" * 72)
            return x1
        x0 = x1

print("=== Phương pháp dây cung ===")
r1 = day_cung(f, d=a, x0=b, epsilon=epsilon)
print(f"\nDây cung : x ≈ {r1:.12f}")

print("\n=== Phương pháp Newton ===")
r2 = newton(f, df, x0=a, epsilon=epsilon)
print(f"Newton   : x ≈ {r2:.12f}")
```

## a. $x^5 - 3x^3 + 2x^2 - x + 5 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt

# ===================== THAY ĐỔI Ở ĐÂY =====================
def f1(x):
    return x**5          # hàm số 1

def f2(x):
    return 3*x**3 - 2*x**2 + x - 5                    # hàm số 2 (xóa nếu không cần)

x_min, x_max = -2.5,-2             # khoảng x
# ===========================================================

x = np.linspace(x_min, x_max, 1000)

plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label=r'$f_1(x)$', color='blue', linewidth=1.5)
plt.plot(x, f2(x), label=r'$f_2(x)$', color='red', linewidth=1.5)

plt.axhline(0, color='black', linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.title(r'Đồ thị $f(x)$', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```
