Viết thuật toán tìm phân tách LU cho ma trận A. Áp dụng cho ma trận vuông A cấp 6 cụ thể

## Phân tách LU (Phương pháp Doolittle)

### Bài toán

Cho ma trận vuông $A$ cấp $n$. Tìm hai ma trận $L$, $U$ sao cho:

$$A = L \cdot U$$

trong đó:
- $L$ — ma trận **tam giác dưới**, đường chéo chính bằng **1**
- $U$ — ma trận **tam giác trên**

$$L = \begin{pmatrix} 1 & & & \\ l_{21} & 1 & & \\ \vdots & \ddots & \ddots & \\ l_{n1} & \cdots & l_{n,n-1} & 1 \end{pmatrix}, \quad U = \begin{pmatrix} u_{11} & u_{12} & \cdots & u_{1n} \\ & u_{22} & \cdots & u_{2n} \\ & & \ddots & \vdots \\ & & & u_{nn} \end{pmatrix}$$

---

### Công thức tính

Với mỗi bước $k = 1, 2, \ldots, n$:

**Bước 1 — Tính hàng $k$ của $U$:**

$$u_{kj} = a_{kj} - \sum_{p=1}^{k-1} l_{kp} \cdot u_{pj}, \quad j = k, k+1, \ldots, n$$

**Bước 2 — Tính cột $k$ của $L$:**

$$l_{ik} = \dfrac{a_{ik} - \displaystyle\sum_{p=1}^{k-1} l_{ip} \cdot u_{pk}}{u_{kk}}, \quad i = k+1, k+2, \ldots, n$$

> ⚠️ Nếu $u_{kk} = 0$ thì phép phân tách không thực hiện được — cần hoán đổi hàng (phân tách LUP).

---

### Thuật toán

**Input:** Ma trận vuông $A$ cấp $n$

**Output:** Ma trận $L$ (tam giác dưới, chéo = 1) và $U$ (tam giác trên) thỏa $A = LU$

```
BEGIN
  L ← I_n        // ma trận đơn vị cấp n
  U ← O_n        // ma trận không cấp n

  FOR k = 1 TO n DO

    // Tính hàng k của U
    FOR j = k TO n DO
      U[k][j] ← A[k][j] − Σ L[k][p] * U[p][j]  với p = 1..k−1
    END FOR

    IF |U[k][k]| < tol THEN
      STOP "Không phân tách được: U[k][k] = 0"
    END IF

    // Tính cột k của L
    FOR i = k+1 TO n DO
      L[i][k] ← ( A[i][k] − Σ L[i][p] * U[p][k]  với p = 1..k−1 ) / U[k][k]
    END FOR

  END FOR

  PRINT L, U
END
```

# Câu 28
Viết thuật toán cho phương pháp phân tách LU giải phương trình AX = B. Áp dụng cho bài toán cụ thể cấp 7

## Giải hệ phương trình $AX = B$ bằng phân tách LU

### Ý tưởng

Từ phân tách $A = LU$, thay vào hệ:

$$AX = B \;\Longrightarrow\; LUX = B$$

Đặt $Y = UX$, ta tách thành **hai hệ tam giác**:

$$\underbrace{LY = B}_{\text{thế xuôi}} \quad \longrightarrow \quad \underbrace{UX = Y}_{\text{thế ngược}}$$

---

### Công thức

**Bước 1 — Phân tách** $A = LU$ (thuật toán Doolittle, xem Câu 27)

**Bước 2 — Thế xuôi** (giải $LY = B$, $L$ tam giác dưới, chéo = 1):

$$y_i = b_i - \sum_{p=1}^{i-1} l_{ip}\, y_p, \quad i = 1, 2, \ldots, n$$

**Bước 3 — Thế ngược** (giải $UX = Y$, $U$ tam giác trên):

$$x_i = \frac{1}{u_{ii}}\!\left(y_i - \sum_{j=i+1}^{n} u_{ij}\, x_j\right), \quad i = n, n{-}1, \ldots, 1$$

---

### Thuật toán

**Input:** Ma trận $A$ cấp $n$, vế phải $B$ (vector cột cấp $n$)

**Output:** Nghiệm $X$ thỏa $AX = B$

```
BEGIN

  // ── Bước 1: Phân tách LU ────────────────────────────────────
  L ← I_n ;  U ← O_n
  FOR k = 1 TO n DO
    FOR j = k TO n DO
      U[k][j] ← A[k][j] − Σ L[k][p]*U[p][j]   (p = 1..k−1)
    END FOR
    IF |U[k][k]| < tol THEN  STOP "A suy biến"
    FOR i = k+1 TO n DO
      L[i][k] ← ( A[i][k] − Σ L[i][p]*U[p][k] (p=1..k−1) ) / U[k][k]
    END FOR
  END FOR

  // ── Bước 2: Thế xuôi  LY = B ────────────────────────────────
  FOR i = 1 TO n DO
    Y[i] ← B[i] − Σ L[i][p]*Y[p]   (p = 1..i−1)
  END FOR

  // ── Bước 3: Thế ngược  UX = Y ───────────────────────────────
  FOR i = n DOWNTO 1 DO
    X[i] ← ( Y[i] − Σ U[i][j]*X[j]  (j = i+1..n) ) / U[i][i]
  END FOR

  PRINT X

END
```

#### Bước 1 — Phân tách $A = LU$

#### Kết quả

**Kiểm tra:** $A \cdot X = B$ ✓


````python
import numpy as np

# ── Hiển thị ma trận / vector ────────────────────────────────────
def _print_mat(M, label):
    n, m = M.shape
    sep = "─" * (12 * m)
    print(f"\n{label}:\n{sep}")
    for row in M:
        print("  " + "  ".join(f"{v:10.4f}" for v in row))
    print(sep)

def _print_vec(v, label):
    print(f"\n{label}:")
    for i, val in enumerate(v):
        print(f"  x{i+1} = {val:10.4f}")


# ── Phân tách LU (Doolittle) ─────────────────────────────────────
def lu_decompose(A):
    n = len(A)
    L, U = np.eye(n), np.zeros((n, n))
    print("=== Bước 1: Phân tách LU ===")
    for k in range(n):
        U[k, k:] = A[k, k:] - L[k, :k] @ U[:k, k:]
        if abs(U[k, k]) < 1e-12:
            raise ValueError(f"Ma trận suy biến tại cột {k+1}")
        L[k+1:, k] = (A[k+1:, k] - L[k+1:, :k] @ U[:k, k]) / U[k, k]

        print(f"\n  ── k = {k+1} ──")
        u_str = "   ".join(f"u[{k+1},{j+1}]={U[k,j]:9.4f}" for j in range(k, n))
        print(f"  Hàng U[{k+1},:] : {u_str}")
        if k < n - 1:
            l_str = "   ".join(f"l[{i+1},{k+1}]={L[i,k]:9.4f}" for i in range(k+1, n))
            print(f"  Cột  L[:,{k+1}] : {l_str}")

    _print_mat(L, "L"); _print_mat(U, "U")
    print(f"\n  L @ U == A : {np.allclose(L @ U, A)}")
    return L, U


# ── Thế xuôi LY = B ──────────────────────────────────────────────
def forward_sub(L, b):
    n = len(b)
    Y = np.zeros(n)
    print("\n=== Bước 2: Thế xuôi LY = B ===")
    for i in range(n):
        Y[i] = b[i] - L[i, :i] @ Y[:i]
        print(f"  y{i+1} = {Y[i]:.4f}")
    return Y


# ── Thế ngược UX = Y ─────────────────────────────────────────────
def backward_sub(U, Y):
    n = len(Y)
    X = np.zeros(n)
    print("\n=== Bước 3: Thế ngược UX = Y ===")
    for i in range(n-1, -1, -1):
        X[i] = (Y[i] - U[i, i+1:] @ X[i+1:]) / U[i, i]
        print(f"  x{i+1} = {X[i]:.4f}")
    return X


# ── Hàm tổng hợp ─────────────────────────────────────────────────
def lu_solve(A, b):
    A, b = np.array(A, dtype=float), np.array(b, dtype=float)
    L, U = lu_decompose(A)
    Y    = forward_sub(L, b)
    X    = backward_sub(U, Y)
    err  = np.max(np.abs(A @ X - b))
    print(f"\n  ||A·X - B||_∞ = {err:.2e}  →  {'ĐÚNG ✓' if err < 1e-8 else 'SAI ✗'}")
    return X


# ── Dữ liệu bài toán cấp 7 ───────────────────────────────────────
A = np.array([
    [0.05953,	0.03186,	0.09466,	0.09514,	0.03478,	0.047,	0.01516,	0.09841,	0.01701,	0.03281],
    [0.06035,	0.07194,	0.09583,	0.08516,	0.00064,	0.05308,	0.08699,	0.06794,	0.04468,	0.06082],
    [0.07307,	0.08444,	0.06385,	0.07472,	0.04107,	0.08514,	0.07277,	0.09001,	0.07051,	0.03304],
    [0.07984,	0.04551,	0.05665,	0.04349,	0.05256,	0.01644,	0.0891,	0.01881,	0.0014,	0.06476],
    [0.05603,	0.0098,	0.07562,	0.04857,	0.01058,	0.03132,	0.06842,	0.04374,	0.00275,	0.09992],
    [0.05876,	0.05356,	0.01616,	0.01209,	0.00999,	0.07356,	0.02366,	0.05048,	0.08056,	0.0216],
    [0.07854,	0.0897,	0.03872,	0.03786,	0.03373,	0.00599,	0.08687,	0.04372,	0.05682,	0.0629],
    [0.07909,	0.03647,	0.09861,	0.04025,	0.03746,	0.00255,	0.04359,	0.06612,	0.04205,	0.02439],
    [0.02533,	0.09595,	0.00065,	0.04671,	0.03464,	0.04578,	0.07115,	0.04019,	0.04568,	0.01673],
    [0.00902,	0.09266,	0.09269,	0.07965,	0.05633,	0.00469,	0.00403,	0.07378,	0.04205,	0.06873]
        ],dtype =float)

b = np.array([29852, 5035, 15243, 4198, 25630, 44127, 18750, 32496, 19828, 9846], dtype=float)


X = lu_solve(np.eye(len(A)) - A, b)
````

# Câu 29
Viết thuật toán tìm phân tách Choleski của ma trận A. Áp dụng cho trường hợp cụ thể với ma trận A cấp 8.

## Phân tách Cholesky

### Bài toán

Cho ma trận vuông $A$ cấp $n$ **đối xứng xác định dương**. Tìm ma trận tam giác dưới $L$ sao cho:

$$A = L \cdot L^T$$

### Điều kiện áp dụng

$A$ phải là ma trận **đối xứng xác định dương** (SPD):
- $A = A^T$
- $\mathbf{x}^T A \mathbf{x} > 0$ với mọi $\mathbf{x} \neq \mathbf{0}$

Tương đương: mọi minor chính của $A$ đều dương (tiêu chuẩn Sylvester).

---

### Công thức

Với mỗi bước $k = 1, 2, \ldots, n$:

**Phần tử đường chéo:**

$$l_{kk} = \sqrt{a_{kk} - \sum_{p=1}^{k-1} l_{kp}^2}$$

**Phần tử phía dưới đường chéo (cột $k$):**

$$l_{ik} = \dfrac{a_{ik} - \displaystyle\sum_{p=1}^{k-1} l_{ip}\, l_{kp}}{l_{kk}}, \quad i = k+1, k+2, \ldots, n$$

---

### Thuật toán

**Input:** Ma trận đối xứng xác định dương $A$ cấp $n$

**Output:** Ma trận tam giác dưới $L$ thỏa $A = LL^T$

```
BEGIN
  L ← O_n

  FOR k = 1 TO n DO

    L[k][k] ← sqrt( A[k][k] − Σ L[k][p]²   với p = 1..k−1 )

    IF L[k][k] ≤ 0 THEN  STOP "A không xác định dương"

    FOR i = k+1 TO n DO
      L[i][k] ← ( A[i][k] − Σ L[i][p]·L[k][p]   với p = 1..k−1 ) / L[k][k]
    END FOR

  END FOR

  PRINT L
END
```

**Kiểm tra:** $L \cdot L^T = A$ ✓

````python
import numpy as np

def _print_mat(M, label):
    n, m = M.shape
    sep = "─" * (12 * m)
    print(f"\n{label}:\n{sep}")
    for row in M:
        print("  " + "  ".join(f"{v:10.4f}" for v in row))
    print(sep)


def cholesky_decompose(A):
    n = len(A)
    L = np.zeros((n, n))
    print("=== Phân tách Cholesky: A = L·Lᵀ ===")
    for k in range(n):
        s = A[k, k] - L[k, :k] @ L[k, :k]
        if s <= 0:
            raise ValueError(f"A không xác định dương tại k={k+1}")
        L[k, k] = np.sqrt(s)
        L[k+1:, k] = (A[k+1:, k] - L[k+1:, :k] @ L[k, :k]) / L[k, k]

        print(f"\n  ── k = {k+1} ──")
        print(f"  l[{k+1},{k+1}] = {L[k,k]:.4f}")
        if k < n - 1:
            l_str = "   ".join(f"l[{i+1},{k+1}]={L[i,k]:9.4f}" for i in range(k+1, n))
            print(f"  Cột L[:,{k+1}] : {l_str}")

    _print_mat(L, "L")
    print(f"\n  L @ Lᵀ == A : {np.allclose(L @ L.T, A)}")
    return L


# ── Dữ liệu ──────────────────────────────────────────────────────
A = np.array([
    [ 4,  2,  0,  0,  0,  0,  0,  0],
    [ 2,  5,  2,  0,  0,  0,  0,  0],
    [ 0,  2, 10,  3,  0,  0,  0,  0],
    [ 0,  0,  3,  5,  2,  0,  0,  0],
    [ 0,  0,  0,  2, 10,  3,  0,  0],
    [ 0,  0,  0,  0,  3,  5,  2,  0],
    [ 0,  0,  0,  0,  0,  2, 17,  4],
    [ 0,  0,  0,  0,  0,  0,  4, 10],
], dtype=float)

L = cholesky_decompose(A)
````

Viết thuật toán cho phương pháp Choleski giải phương trình AX = B. Áp dụng cho bài toán cụ thể cấp 7

## Giải hệ $AX = B$ bằng phân tách Cholesky

### Ý tưởng

Nếu $A$ là ma trận đối xứng xác định dương, phân tách $A = LL^T$ cho phép tách hệ $AX = B$ thành **hai hệ tam giác**:

$$AX = B \;\Longrightarrow\; LL^T X = B$$

Đặt $Y = L^T X$:

$$\underbrace{LY = B}_{\text{thế xuôi}} \quad \longrightarrow \quad \underbrace{L^T X = Y}_{\text{thế ngược}}$$

---

### Công thức

**Bước 2 — Thế xuôi** (giải $LY = B$, $L$ tam giác dưới):

$$y_i = \dfrac{1}{l_{ii}}\!\left(b_i - \sum_{p=1}^{i-1} l_{ip}\, y_p\right), \quad i = 1, 2, \ldots, n$$

**Bước 3 — Thế ngược** (giải $L^T X = Y$, $L^T$ tam giác trên):

$$x_i = \dfrac{1}{l_{ii}}\!\left(y_i - \sum_{j=i+1}^{n} l_{ji}\, x_j\right), \quad i = n, n{-}1, \ldots, 1$$

---

### Thuật toán

**Input:** Ma trận đối xứng xác định dương $A$ cấp $n$, vế phải $B$

**Output:** Nghiệm $X$ thỏa $AX = B$

```
BEGIN

  // ── Bước 1: Phân tách Cholesky ──────────────────────────────
  FOR k = 1 TO n DO
    L[k][k] ← sqrt( A[k][k] − Σ L[k][p]²   (p = 1..k−1) )
    IF L[k][k] ≤ 0 THEN  STOP "A không xác định dương"
    FOR i = k+1 TO n DO
      L[i][k] ← ( A[i][k] − Σ L[i][p]·L[k][p]  (p = 1..k−1) ) / L[k][k]
    END FOR
  END FOR

  // ── Bước 2: Thế xuôi  LY = B ────────────────────────────────
  FOR i = 1 TO n DO
    Y[i] ← ( B[i] − Σ L[i][p]·Y[p]  (p = 1..i−1) ) / L[i][i]
  END FOR

  // ── Bước 3: Thế ngược  LᵀX = Y ─────────────────────────────
  FOR i = n DOWNTO 1 DO
    X[i] ← ( Y[i] − Σ L[j][i]·X[j]  (j = i+1..n) ) / L[i][i]
  END FOR

  PRINT X

END
```

````python
import numpy as np

def _print_mat(M, label):
    n, m = M.shape
    sep = "─" * (12 * m)
    print(f"\n{label}:\n{sep}")
    for row in M:
        print("  " + "  ".join(f"{v:10.4f}" for v in row))
    print(sep)


def cholesky_decompose(A):
    n = len(A)
    L = np.zeros((n, n))
    for k in range(n):
        s = A[k, k] - L[k, :k] @ L[k, :k]
        if s <= 0:
            raise ValueError(f"A không xác định dương tại k={k+1}")
        L[k, k] = np.sqrt(s)
        L[k+1:, k] = (A[k+1:, k] - L[k+1:, :k] @ L[k, :k]) / L[k, k]

        print(f"\n  ── k = {k+1} ──")
        print(f"  l[{k+1},{k+1}] = {L[k,k]:.4f}")
        if k < n - 1:
            l_str = "   ".join(f"l[{i+1},{k+1}]={L[i,k]:9.4f}" for i in range(k+1, n))
            print(f"  Cột L[:,{k+1}] : {l_str}")
    return L


def forward_sub(L, b):
    n = len(b)
    Y = np.zeros(n)
    print("\n=== Bước 2: Thế xuôi LY = B ===")
    for i in range(n):
        Y[i] = (b[i] - L[i, :i] @ Y[:i]) / L[i, i]
        print(f"  y{i+1} = {Y[i]:.4f}")
    return Y


def backward_sub(L, Y):
    n = len(Y)
    X = np.zeros(n)
    print("\n=== Bước 3: Thế ngược LᵀX = Y ===")
    for i in range(n-1, -1, -1):
        X[i] = (Y[i] - L[i+1:, i] @ X[i+1:]) / L[i, i]
        print(f"  x{i+1} = {X[i]:.4f}")
    return X


def cholesky_solve(A, b):
    A, b = np.array(A, dtype=float), np.array(b, dtype=float)
    print("=== Bước 1: Phân tách Cholesky ===")
    L = cholesky_decompose(A)
    _print_mat(L, "\nL")
    print(f"  L @ Lᵀ == A : {np.allclose(L @ L.T, A)}")
    Y = forward_sub(L, b)
    X = backward_sub(L, Y)
    err = np.max(np.abs(A @ X - b))
    print(f"\n  ||A·X - B||_∞ = {err:.2e}  →  {'ĐÚNG ✓' if err < 1e-9 else 'SAI ✗'}")
    return X


# ── Dữ liệu ──────────────────────────────────────────────────────
A = np.array([
    [ 4,  2,  0,  0,  0,  0,  0],
    [ 2, 10,  3,  0,  0,  0,  0],
    [ 0,  3,  5,  2,  0,  0,  0],
    [ 0,  0,  2, 17,  4,  0,  0],
    [ 0,  0,  0,  4,  5,  2,  0],
    [ 0,  0,  0,  0,  2, 10,  3],
    [ 0,  0,  0,  0,  0,  3,  5],
], dtype=float)
b = np.array([6, 15, 10, 23, 11, 15, 8], dtype=float)

X = cholesky_solve(A, b)
````
